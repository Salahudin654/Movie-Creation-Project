2. DATA CLEANING

From The Numbers data that we have up hear, we can see that there`s only six columns that do not have any null values showing the released date, the index column, Movie column containing names, production budget, Domestic gross and Worldwide gross

we will analyse and recommend the best movies from the numbers that have an impact on the audience

```python
#let`s see if any duplicates and drop them incase
df1.duplicated(subset = 'id').sum()
```

5682 is a large number of duplicates considering we only have 5782 rows. Lets explore the data further before making a decision on what to do

The "movies" column as well should not have any duplicates. However, on reviewing the column, we have 84 rows with same Movie name. We can drop this rows to avoid contradicting information

```python
duplicates1=df1.duplicated(subset = 'movie')
duplicates1.sum()
```

```python
#Display a sample of rows that are duplicates in our dataframe_budgets_modified.
duplicate_rows = df1[duplicates1]
duplicate_rows.sample(5)
```

let us now drop the dupplicated rows on the movie column

```python
# drop the duplicate rows in movies
df1.drop_duplicates(subset='movie', inplace=True)
df1
```

```python
#Reconfirm duplicates have beed removed
duplicates2 = df1.duplicated(subset='movie')
duplicates2.sum()
```

let us now add another column called gross_profit that takes in worldwide_gross minus the production_budget. This will give us The profit generated. 

We choosed data from Worldwide because we want to see the Impact of the movies on the world viewers and not just on the Domestic viewers alone

```python
# Convert columns to strings and then remove commas and dollar signs
df1['production_budget'] = df1['production_budget'].astype(str).str.replace(',', '').str.replace('$', '')
df1['domestic_gross'] = df1['domestic_gross'].astype(str).str.replace(',', '').str.replace('$', '')
df1['worldwide_gross'] = df1['worldwide_gross'].astype(str).str.replace(',', '').str.replace('$', '')
```

```python
# convert the relevant columns to floats
df1['production_budget'] = df1['production_budget'].astype(float)
df1['domestic_gross'] = df1['domestic_gross'].astype(float)
df1['worldwide_gross'] = df1['worldwide_gross'].astype(float)
```

```python
#let us see our df1 data after removing the dolar sign
df1
```

```python
# let us get the return on investment column which we are going to call it gross_profit
df1['gross_profit'] = df1['worldwide_gross'] - df1['production_budget']
df1
```

let us now rename our columns (domestic_gross, worldwide_gross and gross_profit) by adding usd to it to show that our values are in usd

```python
# rename the columns to help us identify value is in USD and not any other currency.
df1 = df1.rename(columns={'production_budget': 'production_budget_usd',
                          'domestic_gross': 'domestic_gross_usd',
                          'worldwide_gross': 'worldwide_gross_usd',
                          'gross_profit': 'gross_profit_usd'
                         })
df1
```

3. DATA VISUALIZATION

now that our data is clean and ready for analysis, let as see the top 10 most profitable movies that we can recommend to microsoft 

```python
Top_10_most_profitable = df1.sort_values(by='gross_profit_usd', ascending=False).head(10)
Top_10_most_profitable
```

```python
# Create a bar graph of the top 10 movies
plt.bar(Top_10_most_profitable['movie'], Top_10_most_profitable['gross_profit_usd'])

# set the title and axis labels
plt.title('Movie against Profit generated')
plt.xlabel('Movie Title')
plt.ylabel('Gross profit generated')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the graph
plt.show()
```
