```python
# Your code here - remember to use markdown cells for comments as well!
# importing neccesary modules and libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Now that we have imported the neccesary library we need to read and understand the data that we have and explore on them  one after the other

let Start with Data from IMDb 

```python
# reading the data on the  iMDb file
df = pd.read_csv('ZippedData/tmdb.movies.csv.gz')

#displaying the DataFrame
df.head(10)
```

```python
# this is to get infomation about our data in IMDb file regarding the number of columns and rows and data types of every column
df.info()
```

Now lets explore abit more and see if we have any null values in our data 

```python
#checking for null values
df.isnull().sum()
```

from the above summary we have seen that our data from IMDb has no any missing value and we can now perform other operations on it

```python
# lets drop any duplicates if any
df = df.drop_duplicates()
df
```

Now let us see the Top 10 most popular movies 

```python
Top_10_movies = df.sort_values(by = 'popularity', ascending = False).head(10)
Top_10_movies
```

Now lets see how closely correlated each column is to the other and see there relationships

```python
df.corr()
```

let us now visualize using a  heat map to understand more

3. D

```python
sns.heatmap(df.corr(), annot= True)
plt.show()
```

this is somewhat telling us that vote_count and popularity is quite correlated

```python
# Create a bar graph of the top 10 movies
plt.bar(Top_10_movies['original_title'], Top_10_movies['popularity'])

# set the title and axis labels
plt.title('Movie trends in the market')
plt.xlabel('Movie Title')
plt.ylabel('Popularity')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the graph
plt.show()
```
